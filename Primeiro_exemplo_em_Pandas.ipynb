```python
#Importando biblioteca pandas no python
import pandas as pd

combustíveis_df = pd.read_excel("ca-2021-02.xlsx")
```

```python
#Verificaremos o data frame que acabamos de fazer o upload

print (combustíveis_df)

display(combustíveis_df)
```

```python
#Para exibir somente as 50 primeiras linhas 
display(combustíveis_df.head(50))
```

```python
# Comando para delimitar a quantidade de linhas e colunas
display(combustíveis_df.shape)

#Só a quantidade de colunas
print(combustíveis_df.shape[1])

#Só a quantidade de linhas
print(combustíveis_df.shape[0])
```

```python
#Retorna uma descrição das colunas
print(combustíveis_df.info())
```

```python
# Comando para retornar dados estatísticos do banco de dados
display(combustíveis_df.describe())
```

```python
#Filtragem de apenas uma coluna do df

display(combustíveis_df['Revenda'])
```

```python
# Criar outro df apenas com colunas que quero

ca_df = combustíveis_df[['Revenda', 'Municipio', 'Estado - Sigla', 'Produto', 'Valor de Venda']]

display(ca_df)
```

```python
#Procura a linha especificada, no caso a vigésima segunda

display(ca_df.loc[22])
```

```python
#Retorna uma área especificada no caso da segunda linha até a vigésima segunda
display(ca_df.loc[2:22])
```

```python
#Criar df gas_df contendo:
#apenas 4 colunas (Revenda, Municipio, Produto, Valor de venda)
#somente com combustível sendo GASOLINA
#Exibir na tela

gas_df = ca_df.loc[ca_df['Produto'] == 'GASOLINA']
display(gas_df)
```

```python
#Descobrir o preço da gasolina mais cara
display(gas_df['Valor de Venda'].max())
```

```python
#Exibir o posto e o munícipio da gasolina mais cara
display(gas_df[['Revenda','Municipio','Estado - Sigla', 'Valor de Venda']].max())
```

```python
#DataFrame.loc[] com múltiplas  condições para filtro
#Quais os postos e preços que vendem ETANOL mais próximo da minha cidade
Etanol_SP_df = ca_df.loc[(ca_df['Produto'] == 'ETANOL') & (ca_df['Municipio'] == 'SAO PAULO')]
display(Etanol_SP_df)
```

```python
#Ordenar os dados do df de ETANOL em SP
display(Etanol_SP_df.sort_values(by = 'Valor de Venda'))
```

```python
#Média dos preços de GASOLINA e GASOLINA ADITIVADA no bairro da MOOCA/SP
#1 - BAIRRO E MUNÍCIPIO
display(combustíveis_df.loc[(combustíveis_df['Bairro'] == 'MOOCA') & 
                            (combustíveis_df['Municipio'] == 'SAO PAULO') & 
                            (combustíveis_df['Produto'] == 'GASOLINA') | (combustíveis_df['Produto'] == "GASOLINA ADITIVADA"),
                            ['Valor de Venda']].mean())
```
